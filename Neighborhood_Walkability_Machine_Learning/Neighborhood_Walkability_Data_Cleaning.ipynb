## Creating Walkability Category

```python
def categorize_walkability(x):
    """
    converts the NWI score into a category based on teh following criteria
    """
    if 1.00 <= x <= 5.75:
        return "Least_Walkable"
    elif 5.76 <= x <= 10.50:
        return "Below_Average_Walkable"
    elif 10.51 <= x <= 15.25:
        return "Above_Average_Walkable"
    elif 15.26 <= x <= 20.00:
        return "Most_Walkable"
    else:
        return "Invalid value"
    
    

NWI_top50["NWI_category"] = [categorize_walkability(x) for x in NWI_top50["NatWalkInd"]]



```

```python
def fixing_geoID(dataset):
    """
    iterates through the rows in dataset (specifically designed for the smart location dataset) and compiles a new
    GEOID for each observation based on different column values that can be later used for joining with other datasets.
    """
    for i in dataset.index:
        if len(dataset.at[i, "STATEFP"]) > 1:
            newgeoID = dataset.at[i, "STATEFP"] + dataset.at[i, "COUNTYFP_fix"] + dataset.at[i,"TRACTCE_fix"] + dataset.at[i, "BLKGRPCE"]
            dataset.at[i,'new_geoid'] = newgeoID
    return dataset
    
#print(fixing_geoID(NWI_test))
  
NWI_top50_v2 = fixing_geoID(NWI_top50)


print(NWI_top50_v2[NWI_top50_v2["STATEFP"] == "42"][0:10])
```

```python
def renaming_census_data_var(dataset, old_var_names, new_var_names):
    """
    Function iterates through dataset and renames columns included in the old var name list with the corresponding
    index in the new var names list
    """
    for i in range(len(old_var_names)):
        dataset.rename(columns = {old_var_names[i]: new_var_names[i]}, inplace = True)
    return dataset

def cleaning_census_data_header_fips(dataset):
    """
    function drops the first row in dataset; for census data files the first row is the description of the variable
    function also splits the census data GEO_ID variable into two to isolate the census block groups' unique GEOID
    """
    data = dataset.drop(0).reset_index(drop = True)
    data["CBG_FIPS"] = [x.split("US")[1] for x in data['GEO_ID']]
    return data

def cleaning_census_data_remove_columns(dataset, new_var_names):
    """
    function makes sure that only the renamed and created variables are kept in the dataset
    """
    clean_table_columns = new_var_names + ["CBG_FIPS"]
    df_final = dataset[dataset.columns.intersection(clean_table_columns)]
    return df_final

def cleaning_census_data_files(dataset, old_var_names, new_var_names):
    """
    combines three helper functions above together into one function for ease of data cleaning
    """
    data = renaming_census_data_var(dataset, old_var_names, new_var_names)
    data_v2 = cleaning_census_data_header_fips(data)
    data_clean = cleaning_census_data_remove_columns(data_v2, new_var_names)
    return data_clean

```

```python
ACS_MHI_clean = cleaning_census_data_files(ACS_MHI, ["B19013_001E"], ['Median_HH_Income'])
ACS_MHI_clean.head()

ACS_MedAge_clean = cleaning_census_data_files(ACS_MedAge, ["B01002_001E", "B01002_002E","B01002_003E"], ["Median_Pop_Age", "Median_Male_Age", "Median_Female_Age"])
ACS_MedAge_clean.head()

ACS_Eth_clean = cleaning_census_data_files(ACS_Eth, race_ethnicity, ["Total_Count", "White", "Black", "AIAN", "Asian", "Hawaiian_PacIsl", "Other_Race", "Two_Races"])
ACS_Eth_clean.rename(columns = {"Total_Count": "Race_Total_Count"}, inplace = True)
#print(ACS_Eth_clean.head())

ACS_Commute_clean = cleaning_census_data_files(ACS_Commute, commute_cols, ["Total_Count", "N_Commute_less10", "N_Commute_10to14", "N_Commute_15to19", "N_Commute_20to24", 
                                                                           "N_Commute_25to29", "N_Commute_30to34", "N_Commute_35to44",
                                                                          "N_Commute_45to59", "N_Commute_more60", "N_car", "N_pubtrans", "N_walk", "N_bike_motorcycle"])
ACS_Commute_clean.rename(columns = {"Total_Count": "Commute_Total_Count"}, inplace = True)
#print(ACS_Commute_clean.head())

ACS_Latino_clean = cleaning_census_data_files(ACS_Latino, ["B03003_001E", "B03003_002E", "B03003_003E"], ["Latino_Total_Count", "N_Not_Latino", "N_Latino"])
ACS_Latino_clean.head()

census2020 = cleaning_census_data_files(Census2020, ["P1_001N"], ["2020_Census_PopTotal"])
census2020.head()

ACS_EdAttain_clean = cleaning_census_data_files(ACS_EdAttain, educ_cols, ["Total_Count", "No_School_Comp", "HS_Diploma", "Associates_Deg", "Bachelors_Deg", "Masters_Deg", "Doctorate_Deg"])
ACS_EdAttain_clean.rename(columns = {"Total_Count": "Educ_Total_Count"}, inplace = True)
print(ACS_EdAttain_clean.head())
```

## Merging Datasets 

```python
NWIcopy_v2 = NWI_top50_v2.copy()



def merging_dataset(list_of_datasets, ld_variable, base_dataset, bd_variable):
    """
    function iterates through the list of datasets and joins them to the base dataset.
    join is a left join on the base dataset
    returns a dataset that includes variables from the base dataset and list of datasets
    
    """
    b = base_dataset
    for i in range(len(list_of_datasets)):
        d = list_of_datasets[i]
        merged_data = b.merge(d, left_on = bd_variable, right_on = ld_variable)
        b = merged_data.drop(columns = ld_variable)
    return merged_data
    

census_data = [ACS_MHI_clean, ACS_MedAge_clean, ACS_Eth_clean, ACS_Commute_clean, ACS_Latino_clean, ACS_EdAttain_clean, census2020]  
    

full_data_v2 = merging_dataset(census_data, "CBG_FIPS", NWIcopy_v2, "new_geoid")

print(full_data_v2.shape)

full_clean_data = full_data_v2.drop(columns = ["index", "OBJECTID", "GEOID10", "GEOID20", 'CBG_FIPS', "CBSA", "CBSA_POP", "CBSA_EMP", "CBSA_WRK"])
```

```python
ACS_Commute_clean_test = ACS_Commute_clean.head()

def create_perc_vars(dataset, total_count_col, transforming_cols):
    """
    function iterates through dataset and for each column included in the transforming cols input returns a new variable
    that represents how much the column value is of the total count population. 
    Returns dataset with new columns that contain float values (if total count is greater than 0)
    """
    for i in dataset.index:
        pop_total = int(dataset.at[i, total_count_col])
        if pop_total != 0:
            for j in range(len(transforming_cols)):
                t_col = transforming_cols[j]
                new_col = t_col + "_perc"
                t_val = int(dataset.at[i, t_col])
                dataset.at[i, new_col] =  round((t_val / pop_total),2)
    return dataset

#print(ACS_Commute_clean_test)

#checking data and found one instance where median male age value was nonsensical; replaced it with the median age/
#median female age for the CBG
```
